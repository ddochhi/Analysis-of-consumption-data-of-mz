## 2. 지역별 매출 분석
### - 시군구 단위로 매출 데이터를 분석하여 어떤 지역이 MZ세대의 소비에 가장 큰 영향을 미치는지 확인할 수 있다. 특정 지역에서 인기 있는 업종이나 선호도가 다른 지역과 다른지를 파악한다.

```python
import pandas as pd

data_drop['기준년월'] = data_drop['기준년월'].astype(str)

data_drop['년'] = data_drop['기준년월'].str[:4]
data_drop['월'] = data_drop['기준년월'].str[4:]

data_drop
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data_mz = data_drop[data_drop['MZ']==1]
region_sales = data_mz.groupby('광역시도명')['카드매출금액'].sum()

region_sales.plot(kind='bar',figsize=(10,6))

plt.title('MZ세대의 지역별 매출 분석')
plt.xlabel('지역')
plt.ylabel('매출 금액')

y_min = min(region_sales)
y_max = max(region_sales)
# plt.ylim(y_min,y_max)


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data_nmz = data_drop[data_drop['MZ']==0]
data_mz = data_drop[data_drop['MZ']==1]
region_sales_nmz = data_nmz.groupby('광역시도명')['카드매출금액'].sum()
region_sales_mz = data_mz.groupby('광역시도명')['카드매출금액'].sum()

# region_sales_nmz.plot(kind='bar',figsize=(10,6))
# region_sales_mz.plot(kind='bar',figsize=(10,6))

fig,ax = plt.subplots(figsize = (10,6))

width=0.2
ax.bar(data_drop['광역시도명'].unique().tolist(),region_sales_nmz,width=width,color='blue')
ax.bar([i + width for i in range(len(data_drop['광역시도명'].unique().tolist()))],region_sales_mz,width=width,color='red')

ax.set_title('MZ여부에 따른 지역별 매출 비교 분석')
ax.set_xlabel('지역')
ax.set_ylabel('매출 금액')

ax.text(12,max(region_sales_mz)+190000000000000,'MZ세대의 지역별 매출',color = 'red')
ax.text(12,max(region_sales_nmz),'MZ세대가 아닌 지역별 매출', color = 'blue')

plt.show()

plt.savefig('mz_sale_analyze.png')
```

```python
# 지역별 매출을 업종 분류에 따라 비교하는 코드
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = 'C:/WINDOWS/FONTS/MALGUN.TTF'

font_name = fm.FontProperties(fname = font_path).get_name()

plt.rc('font', family = font_name)

plt.figure(figsize = (20,12))
sns.barplot(x = '광역시도명', y = '카드매출금액', hue = '업종대분류', data=data)

plt.title('광역시도별 업종대분류 카드매출금액 비교',fontsize = 20)
plt.xlabel('광역시도명')
plt.ylabel('카드매출금액')

plt.legend(loc = 'upper right')
plt.show()
```

```python
target_region = '서울'
target_category = '기타'
filtered = data[(data['광역시도명']==target_region)&(data['업종대분류']==target_category)]

filtered['기준년월'] = pd.to_datetime(filtered['기준년월'],format='%Y%m')

plt.figure(figsize=(10,6))
plt.plot(filtered['기준년월'],filtered['카드매출금액'])
plt.title('서울의 업종대분류가 기타인 카드매출 추이')
plt.xlabel('기간')
plt.ylabel('카드매출금액')
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = 'C:/WINDOWS/FONTS/MALGUN.TTF'

font_name = fm.FontProperties(fname = font_path).get_name()

plt.rc('font', family = font_name)

data = data[data['광역시도명']!= '서울']


plt.figure(figsize = (20,12))
sns.barplot(x = '광역시도명', y = '카드매출금액', hue = '업종대분류', data=data)

plt.title('서울을 제외한 광역시도별 업종대분류 카드매출금액 비교',fontsize = 20)
plt.xlabel('광역시도명')
plt.ylabel('카드매출금액')

plt.legend(loc = 'upper right')
plt.show()
```
